```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import os
```

```python
df = pd.read_csv('D:\\Learning\\Udemy\\Deep Learning Foundation\\House_Sales_Regression\\kc_house_data.csv')
pd.set_option('display.max_columns',None)
df.head()
```

```python
plt.scatter(df.price, df.sqft_living)
```

```python
Y = df.price
X = df.sqft_living

m_current=0
b_current=0
epochs=100
learning_rate=0.00000001

N = float(len(Y))

for i in range(epochs):
    y_current= (m_current * X) + b_current
    cost = sum([data**2 for data in (Y - y_current)])/N
    m_gradient = (2/N) * sum(X * (y_current-Y))
    b_gradient = (2/N) *sum(y_current-Y)
    m_current = m_current - (learning_rate * m_gradient)
    b_current = b_current - (learning_rate * b_gradient)
    plt.scatter(X,Y,color = 'red',label = 'Actual points')
    plt.plot(X,y_current,color='blue',label = 'Fitted line')
    plt.legend()
    plt.title('Step ---'+str(i))
    plt.show()
```
