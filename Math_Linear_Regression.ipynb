```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.sort(np.random.random_sample(100)*100)
```

```python
Y = np.sort(np.random.random_sample(100)*100)
```

```python
plt.scatter(X,Y)
```

```python
## Y=mx+c ##
M_current = 0
C_current = 0

N = len(X)

for i in range(10):
    Y_current = M_current * X + C_current
    Error = Y_current - Y
    cost = sum((Error)**2)/N
    M_grad = (2/N) * sum(X * (Y_current - Y))
    C_grad = (2/N) * sum(Y_current - Y)
    M_current = M_current - M_grad * 0.0001
    C_current = C_current - C_grad * 0.0001
    plt.scatter(X,Y,color='red',label='actual')
    plt.plot(X, Y_current,color='blue',label='Fitted line')
    plt.title('Step '+str(i))
    plt.legend()
    plt.show()
    print(cost)
```
