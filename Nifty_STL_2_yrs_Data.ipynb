```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
```

```python
df = pd.read_csv("D:\\Learning\\Time Series Analysis - Financial\\Nifty_2_yrs.csv")
```

```python
df.head()
```

```python
df = df.dropna(how='any')
```

```python
ts = df['Close']
ts.index = df['Date'] 
```

```python
plt.plot(ts)
```

```python
decomposition = sm.tsa.seasonal_decompose(ts,model='additive',freq=4)

plt.title('Time Series Decomposition')
plt.plot(ts.index,ts,label='ts',color='blue')
plt.legend(loc='best')
```

```python
plt.plot(ts.index,decomposition.trend,label='trend',color='red')
plt.legend(loc='best')
```

```python
plt.plot(ts.index,decomposition.seasonal,label='seasonal',color='green')
plt.legend(loc='best')
```

```python
plt.plot(ts.index,decomposition.resid,label='residual',color='orange')
plt.legend(loc='best')
```

```python
decomposition = sm.tsa.seasonal_decompose(ts,model='additive',freq=4)

plt.title('Time Series Decomposition')
plt.subplot(4,1,1)
plt.plot(ts.index,ts,label='ts',color='blue')
plt.legend(loc='best')
plt.subplot(4,1,2)
plt.plot(ts.index,decomposition.trend,label='trend',color='red')
plt.legend(loc='best')
plt.subplot(4,1,3)
plt.plot(ts.index,decomposition.seasonal,label='seasonal',color='green')
plt.legend(loc='best')
plt.subplot(4,1,4)
plt.plot(ts.index,decomposition.resid,label='residual',color='orange')
plt.legend(loc='best')
plt.tight_layout()
```
